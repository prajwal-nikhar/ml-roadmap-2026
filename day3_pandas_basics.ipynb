```python
#Series
import pandas as pd
import numpy as np

s1 = pd.Series([10, 20, 30, 40]) #default integer index
s2 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd']) #custom index
s3 = pd.Series({'Math': 95, 'Biology': 68, 'Social Science': 88, 'English': 95}) #from dict

print("Default index Series:\n:", s1)
print("\nCustom index series:\n", s2)
print("\nSeries from dict:\n", s3)
```

```python
#Acessing elements
print(f"\nAccess s2['b']: {s2['b']}")
print(f"\nAccess s2.iloc[1]: {s2.iloc[1]}")
print(f"\nAccess s3['Maths']: {s3['Math']}")
```

```python
#Series operations (vectorized automatically)
print("\ns2 * 2:\n", s2 * 2)
print("\ns2 > 15\n", s2 > 15)
print("\nFiltered s2:\n",s2[s2>15])
```

```python
#useful series attributes
print("\nValues:", s2.values)
print("Index:", s2.index.tolist())
print("dtype:", s2.dtype)
print("shape", s2.shape)
```

```python
#DataFrame Creation and Exploration
#DataFrame: 2D table like an excel sheet
#Each column is a series sharing the same index
data = {
    'Name':   ['Asha', 'Ravi', 'Meena', 'Karan', 'Priya'],
    'Age':    [22, 25, 23, 28, 24],
    'City':   ['Bangalore', 'Mumbai', 'Delhi', 'Bangalore', 'Chennai'],
    'Marks':  [88, 92, 79, 95, 84],
    'Course': ['CS', 'MBA', 'CS', 'MBA', 'CS']
}
df = pd.DataFrame(data)
df
```

```python
#Exploration methods
print(f'\nShape (rows, columns): {df.shape}')
print(f'\nColumn names: {df.columns.tolist()}')
print(f'\nData types:\n{df.dtypes}')
print(f'\nDescribe (stats from numeric columns):\n{df.describe()}')
print(f'\nHead (first 2 rows): \n{df.head(2)}')
print(f'\nTail (last 2 rows): \n{df.tail(2)}')
print(f'\nInfo: \n{df.info}')
print(f'\nNull check:\n {df.isnull().sum()}')
print(f'\nCheck unique values in Course column: \n{df['Course'].value_counts()}')
```

```python
#key distinction
#loc: label based: uses actual index labels and column Names
#iloc: position-based: uses integer positions (like NumPy 0,1,2...)
df = pd.DataFrame({
    'Name':   ['Asha', 'Ravi', 'Meena', 'Karan', 'Priya'],
    'Age':    [22, 25, 23, 28, 24],
    'City':   ['Bangalore', 'Mumbai', 'Delhi', 'Bangalore', 'Chennai'],
    'Marks':  [88, 92, 79, 95, 84],
    'Course': ['CS', 'MBA', 'CS', 'MBA', 'CS']
}, index = [10, 20, 30, 40, 50])   # custom index to show the difference clearly

print(f'DataFrame with custom index: \n', df)
```

```python
#loc
print('#loc')
print(f'\nSingle row (index=10):\n {df.loc[10]}')
print(f'\nSingle cell: {df.loc[10, 'Name']}')
print(f'\nRow range 10-30, cols Name & Marks:\n {df.loc[10:30, ['Name', 'Marks']]}')
print(f'\nAll rows, 2 columns:\n {df.loc[:, ['Name', 'City']]}')
```

```python
#iloc
print('#iloc')
print(f'\nFirst row (position 0):\n {df.iloc[0]}')
print(f'\nSingle cell (row:0, col:0): {df.iloc[0, 0]}')
print("\nRows 0-2, cols 0-1:\n", df.iloc[0:3, 0:2])                 # note: 0:3 is EXCLUSIVE (like Python slicing)
print("\nLast row:\n", df.iloc[-1])
print("\nEvery other row:\n", df.iloc[::2])
```

```python
# With default index (0,1,2...) loc and iloc feel the same
# but with custom index (10,20,30...) they differ:
print("\nloc[10]: row with INDEX LABEL 10:", df.loc[10, 'Name'])    # Asha
print("\niloc[0]: row at POSITION 0:", df.iloc[0]['Name'])            # also Asha, but for different reason
print("\niloc[10]: would throw IndexError (only 5 rows!)")
```

```python
# Reset to simple index for clarity
df = pd.DataFrame({
    'Name':   ['Asha', 'Ravi', 'Meena', 'Karan', 'Priya'],
    'Age':    [22, 25, 23, 28, 24],
    'City':   ['Bangalore', 'Mumbai', 'Delhi', 'Bangalore', 'Chennai'],
    'Marks':  [88, 92, 79, 95, 84],
    'Course': ['CS', 'MBA', 'CS', 'MBA', 'CS']
})

# --- Single condition ---
print("Marks > 85:\n", df[df['Marks'] > 85])
print("\nCity == Bangalore:\n", df[df['City'] == 'Bangalore'])

# --- Multiple conditions ---
# & = AND, | = OR, ~ = NOT (must use these, not Python's and/or)
print("\nMarks > 80 AND City == Bangalore:\n",
      df[(df['Marks'] > 80) & (df['City'] == 'Bangalore')])

print("\nMarks > 90 OR Course == CS:\n",
      df[(df['Marks'] > 90) | (df['Course'] == 'CS')])

print("\nNOT from Delhi:\n",
      df[~(df['City'] == 'Delhi')])

# --- isin() for multiple value matching ---
print("\nCity in [Bangalore, Mumbai]:\n",
      df[df['City'].isin(['Bangalore', 'Mumbai'])])

# --- between() for range filtering ---
print("\nAge between 22 and 25:\n",
      df[df['Age'].between(22, 25)])

# --- str methods for string filtering ---
print("\nNames starting with 'A' or 'R':\n",
      df[df['Name'].str.startswith(('A', 'R'))])

# --- query() method — more readable syntax ---
print("\nUsing query():\n",
      df.query("Marks > 85 and City == 'Bangalore'"))

# --- Selecting specific columns after filter ---
print("\nNames of students with Marks > 85:\n",
      df[df['Marks'] > 85]['Name'].values)

print("\nName + Marks for Bangalore students:\n",
      df[df['City'] == 'Bangalore'][['Name', 'Marks']])
```
