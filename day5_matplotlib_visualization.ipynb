# Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np

#Line plot - best for trends over time
years = [2018, 2019, 2020, 2021, 2022, 2023]
netflix_titles = [1200, 1800, 2500, 3200, 4100, 5000]
amazon_titles  = [800,  1200, 1900, 2600, 3300, 4200]

plt.figure(figsize=(10, 6))
plt.plot(years, netflix_titles, color='red', marker='o', linewidth=3, label='Netflix')
plt.plot(years, amazon_titles, color='blue', marker='s', linewidth=3, label='Amazon')

plt.title('Streaming Platfrom Growth (2018-2023)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('streaming_platform_growth.png', dpi=200)
plt.show()
```

# Bar Plot

```python
courses  = ['CS', 'MBA', 'Design', 'Marketing', 'Finance']
students = [450, 320, 180, 210, 290]
avg_marks = [85, 78, 82, 74, 80]

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Sort by students descending for vertical bar
sorted_by_students = sorted(zip(students, courses), reverse=True)
students_sorted = [x[0] for x in sorted_by_students]
courses_sorted_v = [x[1] for x in sorted_by_students]

# Vertical bar
axes[0].bar(courses_sorted_v, students_sorted,
            color=['#3498db','#e74c3c','#2ecc71','#f39c12','#9b59b6'],
            edgecolor='black', linewidth=0.8)
axes[0].set_title('Students per Course', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Courses', fontsize=12)
axes[0].set_ylabel('Number of Students', fontsize=12)
for i, v in enumerate(students_sorted):
    axes[0].text(i, v + 6, str(v), ha='center', fontweight='bold')

# Sort by avg_marks descending for horizontal bar
sorted_by_marks = sorted(zip(avg_marks, courses), reverse=True)
avg_marks_sorted = [x[0] for x in sorted_by_marks]
courses_sorted_h = [x[1] for x in sorted_by_marks]

# Horizontal bar
axes[1].barh(courses_sorted_h, avg_marks_sorted,
             color=['#3498db','#e74c3c','#2ecc71','#f39c12','#9b59b6'],
             edgecolor='black', linewidth=0.8)
axes[1].set_title('Average Marks per Course', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Average Marks', fontsize=12)
axes[1].set_xlim(0, 100)   # fixes the cutoff issue
for i, v in enumerate(avg_marks_sorted):
    axes[1].text(v + 0.6, i, str(v), va='center', fontweight='bold')

plt.tight_layout()
plt.savefig('bar_plots.png', dpi=200, bbox_inches='tight')
plt.show()
```

# Scatter Plot

```python
from matplotlib.lines import lineStyles
np.random.seed(42)
study_hours = np.random.uniform(1, 10, 80)
marks = study_hours * 8 + np.random.normal(0, 5, 80)
marks = np.clip(marks, 0, 100)

colors = ['#2ecc71' if m >= 60 else '#e74c3c' for m in marks]

plt.figure(figsize=(9, 6))
scatter = plt.scatter(study_hours, marks, c=colors, alpha=0.75,
                      edgecolors='black', linewidth=0.4, s=80)

plt.title('Study Hours vs Marks', fontsize=14, fontweight='bold')
plt.xlabel('Study Hours per Day', fontsize=12)
plt.ylabel('Marks', fontsize=12)
plt.axhline(y=60, color='orange', linestyle='--', linewidth=1.5, label='Pass threshold (60)')
plt.legend(['Pass threshold (60)', 'Pass (≥60)', 'Fail (<60)'], fontsize=10)
plt.grid(True, alpha=0.3)

#Add correlation
corr = np.corrcoef(study_hours, marks)[0, 1]
plt.annotate(f'Correlation: {corr:.2f}', xy=(0.05, 0.92),
             xycoords='axes fraction', fontsize=11,
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
plt.tight_layout()
plt.savefig('scatter_plot.png', dpi=200)
plt.show()
```

# Histogram

```python
from matplotlib.lines import lineStyles
from dask.array.slicing import colon
from dask.dot import label
from scipy.stats import alpha
from matplotlib.pyplot import sca
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
class_a = np.random.normal(loc=75, scale=10, size=200)
class_b = np.random.normal(loc=65, scale=15, size=200)

plt.figure(figsize=(10, 5))
plt.hist(class_a, bins=20, alpha=0.6, color='#3498db', edgecolor='black', label='Class A')
plt.hist(class_b, bins=20, alpha=0.6, color='#e74c3c', edgecolor='black', label='Class B')

plt.axvline(class_a.mean(), color='blue', linestyle='--', linewidth='1.5',
            label=f'Class A mean: {class_a.mean():.1f}')
plt.axvline(class_b.mean(), color='red', linestyle='--', linewidth='1.5',
            label=f'Class B mean: {class_b.mean():.1f}')

plt.title('Marks Distribution: Class A vs Class B', fontsize=14, fontweight='bold')
plt.xlabel('Marks', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('histogram.png', dpi=150)
plt.show()
```
